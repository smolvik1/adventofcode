```python
valid_str_digits = {
        'one': '1', 
        'two': '2', 
        'three': '3', 
        'four': '4', 
        'five': '5', 
        'six': '6', 
        'seven': '7', 
        'eight': '8', 
        'nine': '9'
    }
```

```python
def replace_spelled_numbers_with_numbers(word: str, valid_dict: dict) -> str:
    match = []

    for key, replacement in valid_dict.items():
        match_index = word.find(key)
        if match_index >= 0:
            match.append((match_index, key, replacement))

    if match:
        min_tuple = min(match, key=lambda x: x[0])
        max_tuple = max(match, key=lambda x: x[0])

        for tuple in [min_tuple, max_tuple]:
            word = word.replace(tuple[1], tuple[2])

    return word
```

```python
data = [replace_spelled_numbers_with_numbers(word, valid_str_digits) for word in data]
print(data)
```

```python
relevant_digits = []

for row in data:

    digit_str=''.join(str(letter) for letter in row if letter.isdigit())
    relevant_digit = int(digit_str[0] + digit_str[-1])
    relevant_digits += relevant_digit

print()
```
