```python
# Load input_sample.txt into a list
with open("input.txt", "r") as f:
    data = f.read().splitlines()

print(f"Loaded {len(data)} lines")
print(data[0:10])
```

```python
def parse_instructions(data: list) -> list[int]:
    instructions = []
    for ins in data:
        move = int(ins[1:])
        if ins[0] == "L":
            instructions.append(-move)
        else:
            instructions.append(move)
    return instructions
```

```python
instructions = parse_instructions(data)
instructions[0:10]
```

```python
def get_new_pos(curr_pos: int, increment: int, max_pos: int = 100) -> int:
    return (curr_pos + increment) % max_pos
```

```python
curr_pos = 50
dial_pos = [curr_pos]
zero_passes = []

for ins in instructions:
    new_pos = get_new_pos(curr_pos, increment=ins)
    dial_pos.append(new_pos)
    curr_pos = new_pos

print(dial_pos)
```

```python
password = dial_pos.count(0)
print(f"The password for part #1: {password}")
```
